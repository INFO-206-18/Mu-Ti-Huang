## 1.1. Simple Edits and Fast-Forward Merging

### Developer A:

If you are Developer A, find the clone URL for this repository.  You can find this on the repo's home page (you may need to hit the back button) by clicking a bright green button that says "clone or download".

In a separate browser window, navigate to the course Jupyterhub and open a terminal.  You will begin in your home directory.

Type `git clone` followed by the clone URL to create an exact copy of this repository on your own file system.  We will call this your local repository.  (Yes, it's technically in the cloud, but you have direct control over it.  Think of the container in the cloud as your own personal computer)

Type "ls" to list the contents of your home directory.  If you cloned the repo correctly, you should now see a folder named "week_01".

Next, use terminal commands to create a folder named `A` in your week_01 folder as shown below.  Create a text file named `dog.txt` inside this folder that contains a single line of poetry about a dog.  The file structure when you're done should look like this.

```
week_01
|---A
    |---dog.txt

```

Using the git commands you've learned, stage all of your changes and then commit them locally.  You should type `git status` frequently along the way to make sure you know what's going on.  At this point, when you type `git status`, you should get the friendly message,

> nothing to commit, working tree clean

Finally, use `git push` to transmit your changes back to the remote repository on github.

### Developer B:

If you are Developer B, wait under Developer A has finished the previous part.

In a separate browser window, navigate to the course JupyterHub and open a terminal.  Clone the repository into your home directory.  You should see the work your partner has already done.  Navigate to the file `dog.txt` and read the poetry your partner has written.

Next, you will add on to the directory structure so that it looks like this:

```
week_01
|---A
|   |---dog.txt
|---B
    |---cat.txt
```

Make cat.txt include a single line of poetry.

Stage all your changes, and then commit them.  Check `git status` a final time to make sure you get the same friendly message as above.  Finally, `git push` to send all your changes back to GitHub.

### Developer A:

You want to see all the work your partner has done.  To get the most recent changes, you should execute a `git pull`.

If you followed the steps exactly, there should not be any merge conflict at this point.  Look at the output and find the words `fast-forward`.  This means that the repository on GitHub is the same as yours, but ahead by one or more commits.  There is no need to merge changes that you committed with changes that your partner comitted.  All that needs to happen is your repository needs to be brought up to date by incorporating your partner's changes. 

Navigate to `cat.txt` and read the poetry your partner has written.

## 1.2. Merge Conflicts

Things get more complicated when two developers work on the same files at the same time.  It's important to understand what happens when you get a merge conflict.

### Developer A and Developer B:

All developers should `git pull` to make sure they are up to date.  (You are probably up to date, but this is a good habit to get into whenever you're starting work).

Add a new line of poetry to dog.txt.  Make sure that your poetry is different than your partner's.

Using your git commands, stage your changes and then commit them.  Do not push to GitHub yet!

### Developer B:

If you are Developer B, go ahead and push your changes to GitHub first.  Your repository is ahead of the one on GitHub, so your job is easy!

### Developer A:

After Developer B pushes to GitHub, you should try to `git push`.  What happens?  Can you explain why this doesn't work?

When there is a merge conflict, you cannot resolve it on GitHub, you have to pull and fix things in your local repository.

Go ahead and git pull.  Look for the word `CONFLICT` in the output and try to read and understand the text that comes after it.

Check what happened to the text in the dog.txt file (For example, you might use `cat dog.txt` if you're in the right directory).  You should see some strange looking lines like `<<<<<<<<` and `========`.  In between these, you should see the poetry you wrote, and also the poetry your partner wrote.

Time to fix the merge conflict.  Open the file in the VI editor (if you're in the right directory, use `vi dog.txt`).  

VI is an archaic, unintuitive text editor, but it tends to pop up when you're using git, so it's important to practice using it a little bit.  To start editing, first press `i` to go into insert mode.  Delete the strange lines and combine the lines of poetry in some way that makes at least a little sense.  Once you are satisfied that your poetry dances off the page and transports the reader into a state of blissful contemplation, hit escape to exit insert mode.  Type `:wq` to save changes and quit VI.

Using your git commands, stage your changes and commit them.

Now try `git push` again.  If you did everything correctly, it should work this time.

### Repeat
